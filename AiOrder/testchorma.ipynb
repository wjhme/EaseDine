```python
from langchain_community.embeddings import SentenceTransformerEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
from langchain_community.vectorstores import FAISS
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
```

```python
embedding = SentenceTransformerEmbeddings(model_name=r"D:\Model\LLM\cache\bge-large-zh-v1.5")
index = faiss.IndexFlatL2(len(embedding.embed_query("hello world")))
vectorstore_mean = FAISS(embedding_function=embedding,docstore= InMemoryDocstore(),index_to_docstore_id={},index=index)
vectorstore_unmean = FAISS(embedding_function=embedding,docstore= InMemoryDocstore(),index_to_docstore_id={},index=index)
```

```python
mess0 = ['色拉轻食, 健康餐']
mess1 = ['主食,饭,盖浇饭']
mean_text = mess0*100 + mess1*100
un_mean = mess1*1000 + mess0*1 
```

```python
vectorstore_mean.add_texts(mean_text)
```

```python
vectorstore_mean.similarity_search("色拉")
```

```python
vectorstore_unmean.add_texts(un_mean)
```

```python
vectorstore_unmean.similarity_search_with_relevance_scores('色拉轻食')
```
